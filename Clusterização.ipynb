```python
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
from matplotlib import pyplot
```

```python
# criando o dataset
X, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)

# definindo o modelo
model = KMeans(n_clusters=2)

# treinando o modelo
model.fit(X)

# avaliando uma nova amostra (predição de cluster)
yhat = model.predict(X)

# recuperando os clusters
clusters = unique(yhat) 

# criando um gráfico para exibir os resultados
for cluster in clusters:

# filtrando as amostras de um mesmo cluster
    row_ix = where(yhat == cluster)

# plotando no gráfico
pyplot.scatter(X[row_ix, 0], X[row_ix, 1])

# exibindo o resultado 
pyplot.show()
```

Affinity Propagation

```python
from sklearn.cluster import AffinityPropagation

# definindo o modelo
modelap = AffinityPropagation(damping=0.9)

# treinando o modelo
modelap.fit(X)

# avaliando uma nova amostra (predição de cluster)
yhatap = modelap.predict(X)

# recuperando os clusters
clusters = unique(yhatap)

# criando um gráfico para exibir os resultados
for cluster in clusters:
    row_ix = where(yhatap == cluster)

# plotando no gráfico
pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

DBSCAN (Clustering Espacial Baseado em Densidade)

```python
from sklearn.cluster import DBSCAN

# definindo o modelo
modeldbscan = DBSCAN(eps=3, min_samples=2)

# treinando o modelo
modeldbscan.fit(X)

# avaliando uma nova amostra (predição de cluster)
yhatdbscan = modeldbscan.fit_predict(X)

# recuperando os clusters
clusters = unique(yhatdbscan)

# criando um gráfico para exibir os resultados
for cluster in clusters:
     row_ix = where(yhatdbscan == cluster)

# plotando no gráfico
pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

Clustering Hierárquico (HC)

```python
from sklearn.cluster import AgglomerativeClustering

# definindo o modelo
modelac = AgglomerativeClustering(affinity='euclidean', n_clusters=3, linkage='single')

# treinando o modelo
modelac.fit(X)

# avaliando uma nova amostra (predição de cluster)
yhatac = modelac.fit_predict(X)

# recuperando os clusters
clusters = unique(yhatac)

# criando um gráfico para exibir os resultados
for cluster in clusters:
  row_ix = where(yhatac == cluster)

# plotando no gráfico
pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```
